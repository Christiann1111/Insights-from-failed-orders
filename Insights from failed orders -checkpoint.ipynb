## Plot the distribution of failed orders by hours. Is there a trend that certain hours have an abnormally high proportion of one category or another? What hours are the biggest fails? How can this be explained?


```python
cancel_system=df[df['order_status_key']==9].groupby('Hour')
cancel_system = cancel_system.order_status_key.count()
cancel_system = pd.DataFrame(cancel_system)
cancel_system
```

```python
cancel_user=df[df['order_status_key']==4].groupby('Hour')
cancel_user = cancel_user.order_status_key.count()
cancel_user = pd.DataFrame(cancel_user)
cancel_user
```

```python
import matplotlib.pyplot as plt


hours = range(1,25)

plt.bar(hours,cancel_user['order_status_key'])
plt.xticks(hours)
plt.ylabel('Cancel')
plt.xlabel('Hours')
plt.title('Users Cancel')



plt.show()

## The users used to cancell so many times over 9 am
```

```python
plt.bar(hours,cancel_system['order_status_key'])
plt.xticks(hours)
plt.ylabel('Cancell')
plt.xlabel('Hours')
plt.title('System cancel')

plt.show()

## The system used to cancel at 9 am
```
